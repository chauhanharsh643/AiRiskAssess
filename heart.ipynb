```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# loading the csv data to a Pandas DataFrame
heart_data = pd.read_csv('heart.csv')
```

```python
# statistical measures about the data
heart_data.describe()
```

```python
# checking the distribution of Target Variable
heart_data['target'].value_counts()
```

```python
X = heart_data.drop(columns='target', axis=1)
Y = heart_data['target']
```

```python
print(len(X))
print(len(Y))
```

```python
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.4, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Logistic Regression Model

```python
log = LogisticRegression(max_iter=1000)
```

```python
# training the LogisticRegression model with Training data
log.fit(X_train, Y_train)
```

```python
# accuracy on training data
X_train_prediction = log.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on Training data : ', training_data_accuracy) 
```

```python
X_test_prediction = log.predict(X_test)
test_data_accuracy_logisticRegression = accuracy_score(X_test_prediction, Y_test)
print('Accuracy on Test data : ', test_data_accuracy_logisticRegression)
```

```python
input_data = (34,0,1,118,210,0,1,192,0,0.7,2,0,2)

# change the input data to a numpy array
input_data_as_numpy_array= np.asarray(input_data)

# reshape the numpy array as we are predicting for only on instance
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

prediction = log.predict(input_data_reshaped)
print(prediction)

if (prediction[0]== 0):
  print('The Person does not have a Heart Disease')
else:
  print('The Person has Heart Disease')
```

Decision Tree Model

```python
from sklearn.svm import SVC
svm = SVC(gamma='auto')
svm.fit(X_train, Y_train)
X_test_prediction = svm.predict(X_test)
test_data_accuracy_svm = accuracy_score(X_test_prediction, Y_test)
print('Accuracy on Test data : ', test_data_accuracy_svm)
```

Random Forest 

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=40)
rf.fit(X_train, Y_train)
```

```python
X_test_prediction = rf.predict(X_test)
test_data_accuracy_randomForest = accuracy_score(X_test_prediction, Y_test)
print('Accuracy on Test data : ', test_data_accuracy_randomForest)
```

Predict Accuracy Fuction

```python
def predictScore(model, X_train, X_test, Y_train, Y_test):
    model.fit(X_train, Y_train)
    X_test_prediction = model.predict(X_test)
    test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
    return test_data_accuracy
```

```python
from sklearn.model_selection import KFold
kf = KFold(n_splits=3)
```

```python
for train_index, test_index in kf.split(heart_data, heart_data.target):
    X__train, X__test = heart_data.iloc[train_index], heart_data.iloc[test_index]
    y__train, y__test = heart_data.target.iloc[train_index], heart_data.target.iloc[test_index]
    
    print(predictScore(LogisticRegression(max_iter=1000), X__train, X__test, y__train, y__test))  
```

StratifiedKFold

```python
import numpy as np
from sklearn.model_selection import StratifiedKFold

folds = StratifiedKFold(n_splits=10)

scores_logistic = []
scores_svm = []
scores_rf = []
scores_knn = []

# Resetting the index to ensure continuous integer indices
heart_data.reset_index(drop=True, inplace=True)

for train_index, test_index in folds.split(heart_data, heart_data.target):
    X__train, X__test = heart_data.iloc[train_index], heart_data.iloc[test_index]
    y__train, y__test = heart_data.target.iloc[train_index], heart_data.target.iloc[test_index]
    
    scores_logistic.append(predictScore(LogisticRegression(max_iter=1000), X_train, X__test, y__train, y__test))  
    scores_svm.append(predictScore(SVC(gamma='auto'), X__train, X__test, y__train, y__test))
    scores_rf.append(predictScore(RandomForestClassifier(n_estimators=40), X__train, X__test, y__train, y__test))
    scores_knn.append(predictScore(NearestCentroid(), X__train, X__test, y__train, y__test))
```
